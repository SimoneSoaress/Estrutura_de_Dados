```python
class ConnectionNode:  # Classe para conexão
    def __init__(self, city, date):
        self.city = city  # Cidade da Conexão
        self.date = date  # Data da Conexão
        self.prox = None  # Próxima Conexão


class Flight:  # Classe do Voo
    def __init__(self, flight_code, origem, destino):
        self.flight_code = flight_code  # Código do Voo
        self.origem = origem  # Origem
        self.destino = destino  # Destino
        self.conexao = None  # Início da Lista

    def add_conexao(self, city, date):
        new_conexao = ConnectionNode(city, date)

        if not self.conexao:
            self.conexao = new_conexao
        else:
            current = self.conexao
            while current.prox:
                current = current.prox
            current.prox = new_conexao

    def Flight_prints(self):
        print(f"Voo: {self.flight_code}")
        print(f"Origem: {self.origem}")
        print(f"Destino: {self.destino}")
        print("Conexões:")

        current = self.conexao
        while current:
            print(f"  - Cidade: {current.city}, Data: {current.date}")
            current = current.prox


def main():
    flights = {}

    while True:
        print("1. Cadastrar voo")
        print("2. Visualizar detalhes do Voo")
        print("3. Sair")
        choice = input("Escolha uma opção: ")

        if choice == "1":
            flight_code = input("Digite o código do voo: ")
            origem = input("Digite a origem: ")
            destino = input("Digite o destino: ")

            flight = Flight(flight_code, origem, destino)

            while True:
                connection_city = input("Digite a cidade da conexão (ou 'fim' para terminar): ")
                if connection_city.lower() == 'fim':
                    break
                connection_date = input("Digite a data da conexão (formato DD/MM/YYYY): ")
                flight.add_conexao(connection_city, connection_date)

            flights[flight_code] = flight
            print("Voo cadastrado com sucesso!")

        elif choice == '2':
            flight_code = input("Digite o código do voo para visualizar: ")
            if flight_code in flights:
                flights[flight_code].Flight_prints()
            else:
                print("Voo não encontrado.")

        elif choice == '3':
            print("Saindo do sistema.")
            break

        else:
            print("Opção inválida, tente novamente.")


if __name__ == "__main__":
    main()
```
