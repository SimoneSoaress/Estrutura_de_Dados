```python
class Estacionamento:
  def __init__(self, capacidade_A, capacidade_B):
    self.capacidade_A = capacidade_A
    self.capacidade_B = capacidade_B
    self.estacionamentoA = [] # Pilha
    self.estacionamentoB = [] # Fila

  def estacionar(self, placa, unidade):
    if unidade == 'A':
      if len(self.estacionamentoA) < self.capacidade_A:
        self.estacionamentoA.append(placa) # Append para a placa ser add na Pilha.
        print(f'Carro{placa} estacionado na unidade A.')
      else:
        print('Estacionamento A cheio, tente novamente mais tarde.')
        print()

    elif unidade == 'B':
      if len(self.estacionamentoB) < self.capacidade_B:
        self.estacionamentoB.append(placa) #Adicionando placa na Fila.
        print(f'Carro{placa} estacionado na unidade B.')
      else:
        print('Estacionamento B cheio.')

  def retirar(self, placa, unidade):
    if unidade == "A":
      if placa in self.estacionamentoA: # Para retirar da pilha, tenho que remover o último carro que entrou.
        while self.estacionamentoA[-1] != placa:
          print(f'Carro {self.estacionamentoA.pop()} teve que ser removido para tirar {placa}.')
        self.estacionamentoA.pop() #para remover o carro desejado.
        print(f'Carro {placa} retirado da unidade A.')
      else:
        print('Carro não encontrado na unidade A.')
        print()

    elif unidade == "B":
      if placa in self.estacionamentoB:
          self.estacionamentoB.remove(placa) #Aqui vou remover o carro da fila
          print(f'Carro {placa} retirado da unidade B.')
      else:
          print('Carro não encontrado na unidade B.')
          print()

  def carros_estacionados(self):
      print('Carros estacionados na unidade A: ', self.estacionamentoA)
      print('Carros estacionados na unidade B: ', self.estacionamentoB)

# Testando o programa de estacionamentos . . .

estacionamento = Estacionamento(capacidade_A=2, capacidade_B=3)

# ADICIONANDO OS CARROS...
estacionamento.estacionar('EAO-5711', 'A')
estacionamento.estacionar('DEV-8940', 'B')
estacionamento.estacionar('FYY-1589', 'A')
estacionamento.estacionar('ODK-6322', 'A')
estacionamento.carros_estacionados()

# REMOVENDO OS CARROS...
estacionamento.retirar('ABC-5714', 'A')
estacionamento.retirar('DEV-8940', 'B')
estacionamento.carros_estacionados()


```
